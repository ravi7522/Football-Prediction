```python
import sqlite3
import pandas as pd
```

```python
# Created connection.
database_path = r"database.sqlite"

conn = sqlite3.connect(database_path)
```

```python
# Creating dataframes from the tbales

df_country = pd.read_sql_query("SELECT * FROM country", conn)
df_league = pd.read_sql_query("SELECT * FROM league", conn)
df_match = pd.read_sql_query("SELECT * FROM match", conn)
df_player = pd.read_sql_query("SELECT * FROM player", conn)
df_player_attributes = pd.read_sql_query("SELECT * FROM player_attributes", conn)
df_team = pd.read_sql_query("SELECT * FROM team", conn)
df_team_attributes = pd.read_sql_query("SELECT * FROM team_attributes", conn)
df_sqlite_sequence = pd.read_sql_query("SELECT * FROM sqlite_sequence", conn)
```

```python
# Concate all the tables in one

# merging country table
df_all = pd.merge(left=df_league, right=df_country, how='outer', left_on='country_id', right_on='id')
df_all = df_all.drop(["id_x", "country_id"], axis=1)
df_all.columns.values[0] = "league_name"
df_all.columns.values[1] = "league_id"
df_all.columns.values[2] = "country_name"

# merging league table
df_all = df_match.merge(df_all, how='inner', left_on='league_id', right_on='league_id')
df_all = df_all.drop(["league_id", "country_id"], axis=1)
df_all.columns.values[0] = "match_id"

df_all
```
